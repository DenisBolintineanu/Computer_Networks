# Filter

**1. Wie lautet der Filter, mit dem Sie über den TCP Port http Verkehr filtern können?**

    tcp.port==80 && http


**2. Erhalten Sie das gleiche Ergebnis wie bei dem Filter HTTP? Erklären Sie ihre Erkenntnis**

    Bei dem Filter HTTP werden alle HTTP-Pakete angezeit unabhängig vom Port.
    Bei dem zusatzt tcp.port==80 werden nur alle HTTP-Pakete über den TCP-Port 80 gesendet wurden.


**3. Was bewirkt der Filter: http && !(udp.port==1900)**

    Dieser Filter zeigt alle HTTP-Pakete an, bei denen es sich nicht um UDP-Pakete auf Port 1900 handelt.
